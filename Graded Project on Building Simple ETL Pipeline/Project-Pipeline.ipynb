```python
#Loading Data: A comprehensive sales report combining information from all sources
import pandas as pd
#reading data from all sources
sales=pd.read_csv("Sales.csv")
customer=pd.read_csv("Customer.csv")
product=pd.read_csv("product.csv")

#merging them into one and writing output to csv file
merged_df = pd.merge(sales,customer[["customer_id","name","email","address"]],on = "customer_id", how = "inner")
merged_df = pd.merge(merged_df,product[["product_id","product_name","category","cost_price"]],on = "product_id", how = "inner")
merged_df.to_csv('sales_report_all_sources.csv')
```

```python
#Loading Data: A summary report with aggregated sales data
import pandas as pd
#adding columns after calculations
merged_df['total_price']=merged_df['quantity']*merged_df['price']
merged_df['total_profit_made']=merged_df['total_price']-merged_df['quantity']*merged_df['cost_price']

#aggregating results and writing it to output csv file
summary=merged_df.groupby('product_name').agg(no_of_sales=('quantity','sum'),
                                              total_sale_amount=('total_price','sum'),
                                              average_sale_price=('price','mean'),
                                              total_profit=('total_profit_made','sum')).reset_index()
summary.to_csv('summary_report_with_aggregated_sales_data.csv')
merged_df
```

```python
#Data Analysis:Calculate total sales and profit by product category.
import pandas as pd
total_sales_profit=merged_df.groupby('category').agg(no_of_sales=('quantity','sum'),
                                                     total_sale_amount=('total_price','sum'),
                                                     total_profit=('total_profit_made','sum')).reset_index()
total_sales_profit
```

```python
#Data Analysis:Identify top-selling products and key customers.
#top_selling_products
top_products=merged_df.groupby('product_name').agg(total_quantity=('quantity','sum'),
                                                   total_sales=('total_price','sum')).sort_values(by='total_sales',ascending=False).reset_index()
#top_products

#Key customers
top_customers=merged_df.groupby('name').agg(total_quantity=('quantity','sum'),
                                            total_spend=('total_price','sum')).sort_values(by='total_spend',ascending=False).reset_index()
top_customers
```

```python
#Data Analysis:Analyze sales trends over time (daily, monthly, quarterly).
#Daily Sales:
merged_df['date']=pd.to_datetime(merged_df['date'],format='%d-%m-%Y')
daily_sales = merged_df.groupby('date')['total_price'].sum().reset_index()
daily_sales
```

```python
#Data Analysis:Analyze sales trends over time (daily, monthly, quarterly).
#Monthly Sales:
merged_df['month'] = merged_df['date'].dt.to_period('M')
monthly_sales = merged_df.groupby('month')['total_price'].sum().reset_index()
monthly_sales['month'] = monthly_sales['month'].dt.to_timestamp()
monthly_sales
```

```python
#Data Analysis:Analyze sales trends over time (daily, monthly, quarterly).
#quarterly Sales:
merged_df['quarter'] = merged_df['date'].dt.to_period('Q')
monthly_sales = merged_df.groupby('quarter')['total_price'].sum().reset_index()
monthly_sales['quarter'] = monthly_sales['quarter'].dt.to_timestamp()
monthly_sales
```
